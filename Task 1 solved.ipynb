### Q8:- Given the first two terms of an Arithmetic Series. Find the nth term of the series. Assume all inputs are provided by the user.

> $S_n = a + T_2 +  T_3 +  T_4 + \dots$ <br><br>
$S_n =$ Arithmetic Series <br>
$a =$ First Term <br>
$T_2 =$ Second \space Term <br>
$d = $ Common Difference <br>
$n = $ Number of Terms <br><br>
$\boxed{T_n = nth \space Term = a + (n - 1)d}$

````python
# 2. Calculate the nth term using the formula: an = a1 + (n-1)*d

n1 = int(input("Enter n1 : "))
n2 = int(input("Enter n2 : "))
n3 = int(input("Enter nth number : "))
d = n2-n1
a = n1+(n3-1)*d
print(a)

````
